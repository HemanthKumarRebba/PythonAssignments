# 4.What is logging in Python?

# 5.What is the significance of the __del__ method in Python?
-  The __del__ method is a special method that is automatically called when an object is about to be destroyed (i.e., when it is garbage-collected).
- It is typically used to release external resources such as closing files, freeing memory, or disconnecting from databases before the object is removed from memory.
- Python’s garbage collector decides when to call it — you do not call it manually in most cases.
