```python
# 6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powerOf2(exp):
  for i in range(exp+1):
    yield (2 ** i )
iter_1 = powerOf2(3)
next(iter_1)
next(iter_1)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def each_line(file_data):
  str_1 = ""
  for i in file_data:
    if i == '\n':
      yield str_1.strip()
      str_1 = ""
    else:
      str_1 = str_1 + i
file_data = """Hemanth kumar rebba
Lokesh Kumar
"""
gen_fun = each_line(file_data)
next(gen_fun)
next(gen_fun)
```

```python
#8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

list_1 = [(1, 2, 3), (2, 1, 3), (1, 3, 2)]
list_2 = sorted(list_1, key = lambda x : x[1])
print(list_2)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
list_1 = [20, 15, 30]
list_2 = list(map(lambda x : x * 9/5 + 32, list_1))
print(list_2)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
str_1 = "pwskills"
list_1 = list(str_1)
list_2 = list(filter(lambda x: not(x == 'a') and not(x == 'e') and not(x == 'i') and not(x == 'o') and not(x == 'u'), list_1))
print(''.join(list_2))
```
