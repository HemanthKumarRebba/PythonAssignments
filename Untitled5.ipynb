```python
# Original list of strings

str_list = ['1', '2', '3', '4', '5']

# Convert each string element to integer
int_list = [int(item) for item in str_list]

# Convert each integer to float
float_list = [float(item) for item in int_list]

# Convert each float to string again
string_list = [str(item) for item in float_list]

# Print results
print("Original string list:", str_list)
print("After casting to int:", int_list)
print("After casting to float:", float_list)
print("After casting back to string:", string_list)
```

```python
#Write a Python program to find the sum of all even numbers between 1 and 50.
sum = 0
for i in range(1, 50):
  if i % 2 == 0:
    sum = sum + i
print(sum)
```

```python
# Write a program to reverse a string using a while loop.
str_1 = input("Enter the String:")
str_2 = ""
for i in range(len(str_1)-1, -1, -1):
  str_2 += str_1[i]
print(str_2)
```

```python
#Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter the number:"))
fact = 1
while(num != 0):
  fact *= num
  num -= 1
print(fact)
```
